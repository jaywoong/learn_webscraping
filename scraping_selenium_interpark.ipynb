```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('http://ticket.interpark.com/Contents/Ranking?smid1=s_menu&smid2=performance')
```

```python
html = browser.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tags = soup.select('div.items > ul > li')
```

```python
tags[0]
```

```python
tag = tags[0]
```

```python
name = tag.select('a.prdName')
name = name[0].text.strip()
name
```

```python
duration = tag.select('a.prdDuration')
duration[0].text.strip()
```

```python
contents = list()
for tag in tags:
    name = tag.select('a.prdName')
    duration = tag.select('a.prdDuration')
    play = name[0].text.strip(),duration[0].text.strip()
    contents.append(play)
    
contents
```

```python
import pandas as pd
```

```python
pd_data = pd.DataFrame(contents, columns = ['Name', 'Duration'])
pd_data
```

```python
pd_data.to_excel('./saves/interpark_scraping.xlsx', index = False)
```
